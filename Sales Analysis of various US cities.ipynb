# Question 2. Which city had the highest number of sales?

```python
df.head()
```

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1])
```

```python
df
```

```python
df.City.unique()
```

```python
def get_state(x):
    return x.split(',')[2]
df['State'] = df['Purchase Address'].apply(lambda x: get_state(x))
```

```python
df['StateReq'] = df['State'].apply(lambda x : x.split(' ')[1])
```

```python
df
```

```python
df['Cityreq'] = df['City'] + df['StateReq']
```

```python
df
```

```python
df.Cityreq.unique()
```

```python
res = df.groupby('Cityreq').sum()
res
```

```python
df.groupby('Cityreq').sum().max()
```

```python
import matplotlib.pyplot as plt
cities = [Cityreq for Cityreq,df in df.groupby('Cityreq')]
plt.bar(cities,res['Sales'])
plt.ylabel("Sales in US Dollars")
plt.xlabel("Cities")
plt.xticks(cities, rotation = 'vertical', size=8)            # San Francisco have the highest sales
plt.figure(figsize=(10,10))
```

# Question 4. What products are most often sold together?

```python
df.head()
```

```python
reqdf = df[df['Order ID'].duplicated(keep=False)==True]
reqdf
```

```python
reqdf.groupby('Order ID')['Product']
```

```python
reqdf['Grouped'] = reqdf.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
```

```python
reqdf
```

```python
reqdf[['Order ID','Grouped']]
```

```python
df2 = reqdf[['Order ID','Grouped']].drop_duplicates()
df2
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key,value)
```

```python
product_group = df.groupby('Product')
quantity_ordered = product_group['Quantity'].sum()

keys = [Product for Product, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.ylabel("Quantity ordered")
plt.xlabel("Products")
plt.show()                                           # Batteries are sold the most
```

```python
df.dtypes
```

```python
df['Price of each'] = df['Price Each'].astype(float)
```

```python
df
```

```python
prices = df.groupby('Product')['Price of each'].mean()
```

```python
#plotting the graph to show the corelation between quantity ordered and price 
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
